```python
import requests
import json
```

```python
sentences = [
    "This is a test sentence.",
    "This is another test sentence.",
    "This is a third test sentence."
]

ENDPOINT = "http://localhost:5000"
```

```python
resp = requests.get(f"{ENDPOINT}/health")


if resp.status_code == 200:
    print("Health check passed.")

print(resp.json())
```

```python
resp = requests.get(f"{ENDPOINT}/models")

print(json.dumps(resp.json(), indent=2))
```

```python
resp = requests.get(f"{ENDPOINT}/models/sentence-transformers_all-mpnet-base-v2")

print(json.dumps(resp.json(), indent=2))
```

```python
resp = requests.post(f"{ENDPOINT}/models/sentence-transformers_all-mpnet-base-v2/embed", json=sentences)

print(json.dumps(resp.json(), indent=2))
```
